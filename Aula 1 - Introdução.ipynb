## 1. O conceito de vibração

Qualquer movimento oscilatório pode ser denominado vibração. Ou seja, Vibração pode ser definida como qualquer movimento que se repete após um certo intervalo de **tempo**.

Para Rao(2008), a teoria das vibrações trata do estudo de movimentos oscilatórios de corpos e as forças associadas a eles.

Um exemplo oscilatório clássico é o pêndulo simples, em que a massa, presa a um fio indeformável, oscila entre amplitudes constantes (despreza as forças dissipativas). A figura abaixo ilustra o movimento de um pêndulo simples.
<img src="Figuras/pendulosimples.png" width="500">

Como o Pêndulo simples se trata de um **movimento harmônico simples**, seu movimento pode ser descrito pela seguinte equação:
\begin{align}
X = A.\cos{(\phi_0 + \omega t)}
\end{align}
Onde $\omega$ é a velocidade angular do sistema$(\frac{rad}{s})$, que, a partir de agora, será tratada como a **Frequência natural de vibração** do sistema. É válido lembrar que, para trabalhar com a frequência em Hertz, basta utilizar a seguinte expressão:
\begin{align}
\omega = 2\pi f
\end{align}

O script abaixo ilustra a curva Posição-tempo para um pêndulo simples.

```python
A = 2                         #Amplitude do movimento [m]
phi = 0                       #Fase inicial do sistema (Ângulo inicial)
w = 1                         #Frequência natural do sistema(rad/s)
t = np.linspace(0,50,1000)    #Usando a Numpy para discretizar o tempo
X = A*np.cos(phi+w*t)         #Montando a equação
plt.figure(1,figsize=(6,4))   #Plotando a figura e definindo sua numeração e tamanho
plt.plot(t,X,'b')             # plotando os eixos x e y
plt.xlabel("tempo(s)"); plt.ylabel("deslocamento(m)")        #definindo o nome dos eixos
plt.grid(True)                # Plotando a grade
```

## 2. Partes de um sistema oscilatório

Um sistema oscilatório é constituido por massa, mola e amortecedor, sendo cada um associado a um tipo de energia no sistema.
<img src="Figuras/sistema.PNG" width="400">
* A massa é reponsável pela reprentação da parcela cinética da energia, estando associada a componente **inercial** do sistema
* A mola é responsável pela parcela de energia potencial (restitutiva) do sistema, sendo ela a representação da **rigidez** do sistema
* O amortecedor é responsável pela parcela de energia **dissipada** no sistema, podendo este ser devido a atrito, viscosidade ou histerese do sistema.

No exemplo anterior, o pêndulo simples livre de dissipação de energia ( sem amortecedor), visto que sua amplitude permanece constante durante todo o movimento. O script abaixo ilustra o que aconteceria caso houvesse amortecimento no sistema.

```python
X_a = np.exp(-0.05*t*w)*X
plt.figure(1,figsize=(6,4))   #Plotando a figura e definindo sua numeração e tamanho
plt.plot(t,X_a,'b')             # plotando os eixos x e y
plt.xlabel("tempo(s)"); plt.ylabel("deslocamento(m)")        #definindo o nome dos eixos
plt.grid(True)     
```

Quanto a força externa, as vibrações podem ser classificadas como **livres** ou **forçadas**, já quanto a presença de amortecedor, são classificadas como **amortecidas** ou **não amortcidas**. Se o comportamento dos componentes do sistema seguir uma regra linear, será considerado um sistema **linear**, caso contrário, será denominado **não linear**. A vibração **determinística** é aquela que pode ser representada por uma equação, como mostrado nos exemplos anteriores com os Pêndulos com e sem amortecimento. Já a **aleatória** precisa de um tratamento estatístico para poder ser representada. O script abaixo ilustra um movimento aleatório.
