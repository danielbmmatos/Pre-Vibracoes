### Exercício 1:

Representar a viga por um sistema de um grau de liberdade, apresentando os valores de massa equivalente, rigidez equivalente e frequência natural

```python
# DADOS BÁSICOS
E  = 20e9
z  = 0.01
l  = 9
b  = 0.85
h  = 0.25
A  = b*h
I  = b*h**3/12
mu = A * 2500
#Calculo da rigidez e massa equivalente
k  = 48*E*I/(l**3)
m  = 17*mu*l/35
# Calculo da frequência natural
wn = np.sqrt(k/m)
print('Rigidez equivalente: {0:.4f}N/m'.format(k))
print('Massa equivalente: {0:.4f}kg'.format(m))
print('Frequência natural: {0:.4f}rad/s'.format(wn))
```

### Exercício 2:

Calcular a frequência natural amortecida e a resposta máxima deste sistema a uma velocidade inicial de 2 m/s

```python
wd = wn*np.sqrt(1-z**2)
print(wd)
t = np.linspace(0,100,10000)
u0 = 0
v0 = 2
u = np.exp(-z*wn*t)*(u0*np.cos(wd*t)+(v0+z*wn*u0)*np.sin(wd*t)/wd)

plt.figure(1,figsize = (8,4))
plt.plot(t,u*1000,'b')
plt.xlabel('tempo(s)') ; plt.ylabel('deslocamento(mm)')
plt.grid(True)
plt.xlim(0,4)
print(np.max(u))
```

### Exercício 5:

Estimar o valor r.m.s e o deslocamento de pico do sistema(considere a força média de 10kN) quando este está submetido a um ruído branco em banda como ilustra a figura abaixo:
<img src="Figuras/ex4.jpg" width="500">


```python
w = np.linspace(2*np.pi,10*np.pi,1000)
dw = 4*np.pi/1000
siga = 0.2*9.81
siga_2 = siga**2
s0     = siga_2/np.pi/4
um     = 10000/k

Sa     =np.ones(len(w))*s0 


b = w/wn
H = 1/(wn**4*((1-b**2)**2 + (2*z*b)**2))

Su     = H*Sa

sigu_2 = np.trapz(Su,w)
sigu   =np.sqrt(sigu_2)
print(um)
print(np.sqrt(um**2+sigu**2))
print(um + 4*sigu)
plt.semilogy(w,H)
plt.grid(True)
```
